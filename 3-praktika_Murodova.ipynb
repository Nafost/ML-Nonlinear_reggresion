```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

np.random.seed(0)
X = np.sort(5 * np.random.rand(20, 1), axis=0)  
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])  

poly = PolynomialFeatures(degree=3)  # 3-darajali polynomial
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

X_test = np.linspace(0, 5, 100).reshape(-1, 1)
X_test_poly = poly.transform(X_test)
y_pred = model.predict(X_test_poly)

plt.scatter(X, y, color='red', label="Haqiqiy ma'lumotlar")
plt.plot(X_test, y_pred, color='blue', label="Polynomial Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title("No-chiziqli regressiya (Polynomial Regression)")
plt.show()
```
